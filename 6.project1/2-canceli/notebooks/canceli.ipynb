    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.

```python
to_be_dropped_cols = ['BillID', 'TicketID', 'UserID', 'HashPassportNumber_p', 'HashEmail',
                        'BuyerMobile', 'NationalCode', 'VehicleType']

summarized_train_data = train_data.drop(columns=to_be_dropped_cols+['CancelTime'], axis=1)
summarized_test_data = test_data.drop(columns=to_be_dropped_cols, axis=1)

x_train = summarized_train_data.drop(columns=['Cancel'], axis=1)
y_train = summarized_train_data['Cancel']
```

```python
dates_delta_train = (pd.to_datetime(x_train['DepartureTime'], format='%Y-%m-%d %H:%M:%S') - 
                pd.to_datetime(x_train['Created'], format='%Y-%m-%d %H:%M:%S'))

dates_delta_test = (pd.to_datetime(summarized_test_data['DepartureTime'], format='%Y-%m-%d %H:%M:%S') - 
                pd.to_datetime(summarized_test_data['Created'], format='%Y-%m-%d %H:%M:%S'))

dates_delta_train = dates_delta_train.dt.total_seconds()

dates_delta_test = dates_delta_test.dt.total_seconds()

x_train['Departure - Created'] = dates_delta_train

summarized_test_data['Departure - Created'] = dates_delta_test

x_train.drop(columns=['DepartureTime', 'Created'], inplace=True)
summarized_test_data.drop(columns=['DepartureTime', 'Created'], inplace=True)
```

```python
x_train.describe()
```

```python
summarized_test_data.describe()
```

```python
price_outliers = x_train[(x_train['Price'] <= 0)]['Price']

s1_price = sum(x_train['Price'])
s2_price = sum(price_outliers)

mean_without_price_outliers = (s1_price - s2_price) // (x_train['Price'].shape[0] - price_outliers.shape[0])

x_train.loc[(x_train['Price'] <= 0), 'Price'] = mean_without_price_outliers

discount_outliers = summarized_test_data[(summarized_test_data['CouponDiscount'] < 0)]['CouponDiscount']

s1_discount = sum(summarized_test_data['CouponDiscount'])
s2_discount = sum(discount_outliers)

mean_without_discount_outliers = ((s1_discount - s2_discount) // 
                                    (summarized_test_data['CouponDiscount'].shape[0] - discount_outliers.shape[0]))

summarized_test_data.loc[(summarized_test_data['CouponDiscount'] < 0), 'CouponDiscount'] = mean_without_discount_outliers
```

```python
binary_cols = ['Male', 'VehicleClass']
cat_cols = ['From', 'To']
nominal_cols = ['Vehicle', 'TripReason']
```

```python
binary_mapping = {True: 1, False: 0}

for col in binary_cols:
    x_train[col] = x_train[col].map(binary_mapping)
    summarized_test_data[col] = summarized_test_data[col].map(binary_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

x_train['train'] = 1
summarized_test_data['train'] = 0

combined_df = pd.concat([x_train, summarized_test_data])

for col in cat_cols:
    le.fit(combined_df[col])
    x_train[col] = le.transform(x_train[col])
    summarized_test_data[col] = le.transform(summarized_test_data[col])

x_train.drop(columns=['train'], inplace=True)
summarized_test_data.drop(columns=['train'], inplace=True)
```

```python
nominal_train = x_train[(col for col in nominal_cols)]
nominal_test = summarized_test_data[(col for col in nominal_cols)]

nominal_train = pd.get_dummies(nominal_train, prefix=nominal_cols)
nominal_test = pd.get_dummies(nominal_test, prefix=nominal_cols)

x_train = x_train.join(nominal_train)
summarized_test_data = summarized_test_data.join(nominal_test)

x_train.drop(nominal_cols, axis=1, inplace=True)
summarized_test_data.drop(nominal_cols, axis=1, inplace=True)
```

```python
# target = x_train[(x_train['VehicleClass'].isna())]
train_col = x_train[~(x_train['VehicleClass'].isna())]

train_col_x = train_col.drop(columns=['VehicleClass'], axis=1)
train_col_y = train_col['VehicleClass']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import f1_score

g_s = GridSearchCV(RandomForestClassifier(), 
                {'n_estimators': [50, 75, 100], 'max_depth': [5, 10, 15, 20],
                    'min_samples_split':[1, 2, 3, 4]},
                    scoring='f1')

g_s.fit(train_col_x, train_col_y)

g_s.best_params_
```

```python
model = RandomForestClassifier(n_estimators = 100, min_samples_split = 1, max_depth = 20)

model.fit(train_col_x, train_col_y)

train_col_y_pred = model.predict(train_col_x)

f1_score(train_col_y, train_col_y_pred)
```

```python
x_train['VehicleClass'] = x_train['VehicleClass'].fillna(
    pd.Series(model.predict(x_train.drop('VehicleClass', axis=1))))
```

```python
summarized_test_data['VehicleClass'] = summarized_test_data['VehicleClass'].fillna(
    pd.Series(model.predict(summarized_test_data.drop('VehicleClass', axis=1)))
)
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    معیاری که برای ارزیابی عملکرد مدل انتخاب کرده‌ایم، <code>f1_score</code> نام دارد.
    <br>
    این معیار، سنجه ارزیابی کیفیت مدل شماست. به عبارت بهتر در سامانه داوری هم از همین معیار برای نمره‌دهی استفاده شده است.
    <br>
    پیشنهاد می‌شود با توجه به این معیار، عملکرد مدل خود را بر روی مجموعه داده آموزش یا اعتبارسنجی ارزیابی کنید.
</font>

```python
from sklearn.metrics import  f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

model = RandomForestClassifier(n_estimators=100, min_samples_split=3, max_depth=25)
model.fit(x_train, y_train)

y_train_pred = model.predict(x_train)
y_val_pred = model.predict(x_val)

print(f'training f1 score: {f1_score(y_train, y_train_pred)}')
print(f'validation f1 score: {f1_score(y_val, y_val_pred)}')
```

```python
print(classification_report(y_train, y_train_pred))
print(classification_report(y_val, y_val_pred))
```

```python
cm_train = confusion_matrix(y_train, y_train_pred)
cm_val = confusion_matrix(y_val, y_val_pred)

ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=[False, True]).plot()
ConfusionMatrixDisplay(confusion_matrix=cm_val, display_labels=[False, True]).plot()
```
